```python
import cv2
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('G:\wolf.jpeg')

# Check if the image was loaded successfully
if img is None:
    print("Error: Image not loaded or does not exist.")
else:
    # Convert color space
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Display the image
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.show()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray)
```

```python
img_invert = cv2.bitwise_not(img_gray)

plt.imshow(img_invert)
```

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)

plt.imshow(img_smoothing)
```

```python
inverted_blurred = 255 - img_smoothing
pencil_sketch = cv2.divide(img_gray, inverted_blurred, scale=256.0)

     

plt.imshow(pencil_sketch, cmap='gray')  # Specify cmap='gray' for grayscale images
plt.axis('off')  # Turn off axis labels
plt.title('Pencil Sketch')
plt.show()
     
```

```python
plt.figure(figsize=(10, 5))

# Plot the original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title('Original Image')
plt.axis('off')

# Plot the pencil sketch
plt.subplot(1, 2, 2)
plt.imshow(pencil_sketch, cmap='gray')
plt.title('Pencil Sketch')
plt.axis('off')

# Adjust layout
plt.tight_layout()

# Show the figure
plt.show()
```
